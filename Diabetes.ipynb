# EDA and Visualizations

#### Heat Map 

```python
sns.heatmap(df.corr(), annot = True,cmap="coolwarm")
plt.title("Correlation Heat map")
plt.show()
```

#### Diabetes Outcome Count

```python
sns.countplot(x='Outcome', data=df)
plt.title("Diabetes Outcome Count")
plt.xlabel("Outcome (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

#### Histograms 

```python
cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'Age']
for col in cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

#### Outcome Boxplots 

```python
for col in cols:
    plt.figure()
    sns.boxplot(x='Outcome', y=col, data=df)
    plt.title(f"{col} by Diabetes Outcome")
    plt.show()
```

#### Relationships between Attributes

```python
sns.pairplot(df[['Glucose', 'BMI', 'Age', 'Outcome']], hue='Outcome')
plt.show()
```

#### BMI vs Age by Outcome

```python
sns.scatterplot(x='Age', y='BMI', hue='Outcome', data=df)
plt.title("BMI vs Age by Outcome")
plt.show()
```
